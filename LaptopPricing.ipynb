```python
## import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python

#download the dataset

def download(url, fileName):
    response = requests.get(url)
    if(response.status_code == 200):
        with open(fileName, "wb") as f:
            f.write(response.content)
```

```python
file_path= "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod1.csv"
```

```python
download(file_path, "laptops.csv")
file_name = "laptops.csv"
```

```python
# we are going to create a datafram as df using the laptops.csv

df = pd.read_csv(file_name)
df.head()
```

```python
#we get the summary about dataset
df.info()
```

```python
## rounded the column screensize cm to nearest two decimal places.
## consider as a dataframe

df[['Screen_Size_cm']] = df[['Screen_Size_cm']].astype("float")
df[['Screen_Size_cm']] = np.round(df[['Screen_Size_cm']], 2)
df['Screen_Size_cm'].head()
```

# Task 1: Finding Missing Values

<p> Evaluate the data set for missing data </p>

```python
df.isnull().sum()
```

```python
## Alnternate way
missing_data = df.isnull()
print(missing_data.head())

for colum in missing_data.columns.values.tolist():
    print(missing_data[colum].value_counts())
    print("")
```

## Task 2: Replace with Mean
 <p> weight can be replace with the mean of the column data </p>

```python
df['Weight_kg'].dtype
```

```python
mean_weights = np.round(df['Weight_kg'].mean(axis=0), 2)
print("Mean of weight_kg column: " , mean_weights)
```

```python
df['Weight_kg'] = df['Weight_kg'].fillna(mean_weights)
print(df['Weight_kg'].head())

df['Weight_kg'].isnull().sum()
```

### Replace with the most frequent value
Missing values in attributes that have categorical data are best replaced using the most frequent value. We note that values in "Screen_Size_cm" attribute are categorical in nature, and some values are missing. Therefore, write a code to replace the missing values of Screen Size with the most frequent value of the attribute.
